# In-Class Assignment 1


```python
# Income Statement Program

# Get sales input from user
sales = float(input("Enter sales ($): "))

# Calculate Cost of Goods Sold (40% of Sales)
cost_of_goods_sold = sales * 0.40

# Calculate Gross Profit
gross_profit = sales - cost_of_goods_sold

# Get tax rate from user
tax_rate = float(input("Enter tax rate (%): "))

# Calculate Taxes and Net Income
taxes = gross_profit * (tax_rate / 100)
net_income = gross_profit - taxes

# Display the Income Statement
print("\n Income Statement ")
print(f"Sales:              ${sales:,.2f}")
print(f"Cost of Goods Sold: ${cost_of_goods_sold:,.2f}")
print(f"Gross Profit:       ${gross_profit:,.2f}")
print(f"Taxes:              ${taxes:,.2f}")
print(f"Net Income:         ${net_income:,.2f}")
```

# In-Class Assignment 2

```python
# Simple Numerical Calculator

# Get user inputs
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
operator = input("Enter an operator (+, -, *, /): ")

# Perform calculation
if operator == "+":
    result = num1 + num2
elif operator == "-":
    result = num1 - num2
elif operator == "*":
    result = num1 * num2
elif operator == "/":
    if num2 != 0:
        result = num1 / num2
    else:
        print("Error: Division by zero is not allowed.")
        exit()
else:
    print("Invalid operator.")
    exit()

# Print result
print("Result:", result)
```

```python
import math

ticket_price = 10
base_attendees = 20
fixed_cost = 200

print("Ad$   Profit")

for ad in range(0, 201, 25):
    # Calculate additional attendees
    additional = 2 * math.sqrt(ad)

    # Total attendees
    total_attendees = base_attendees + additional

    # Revenue
    revenue = ticket_price * total_attendees

    # Profit
    profit = revenue - ad - fixed_cost

    # Print results
    print(f"{ad:.1f}  {profit:.1f}")
```

```python
base_fee = 5
hourly_rate = 2.50
min_fee = 10
max_fee = 20

hours = 1

while hours <= 8:
    fee = base_fee + hourly_rate * hours

    # Apply minimum and maximum limits
    if fee < min_fee:
        fee = min_fee
    elif fee > max_fee:
        fee = max_fee

    print(hours, fee)

    hours += 1
```

```python
import math

ticket_price = 10
base_attendees = 20
fixed_cost = 200

print("Ad$   Profit")

ad = 0

while ad <= 200:
    # Calculate additional attendees
    additional = 2 * math.sqrt(ad)

    # Total attendees
    total_attendees = base_attendees + additional

    # Revenue
    revenue = ticket_price * total_attendees

    # Profit
    profit = revenue - ad - fixed_cost

    # Print results
    print(f"{ad:.1f}  {profit:.1f}")

    ad += 25
```

# In-Class Assignment 4

```python
import random   # MUST be at the top

play = input("Do you want to play Rock-Paper-Scissors? (yes/no): ").lower()

while play == "yes":

    user_move = input("Choose rock, paper, or scissors: ").lower()

    # Random computer choice (this IS randomized)
    computer_move = random.choice(["rock", "paper", "scissors"])

    print("Computer chose:", computer_move)

    if user_move == computer_move:
        print("It's a tie!")

    elif (user_move == "rock" and computer_move == "scissors") or \
         (user_move == "paper" and computer_move == "rock") or \
         (user_move == "scissors" and computer_move == "paper"):
        print("You win!")

    elif user_move in ["rock", "paper", "scissors"]:
        print("Computer wins!")

    else:
        print("Invalid choice.")

    play = input("Do you want to play again? (yes/no): ").lower()

print("Thanks for playing!")
```

# In-Class Assignment 5

```python
numbers = []   # start with an empty list

while True:
    user_input = float(input("Enter a number (0 to stop): "))

    if user_input == 0:
        break   # stop the loop when user enters 0

    numbers.append(user_input)  # add number to the list

# Sort the list in descending order
numbers.sort(reverse=True)

# Display the result
print("Numbers in descending order:")
print(numbers)
```

```python
# Create the dictionary with product IDs and prices
products = {
    "P-01": 9.95,
    "P-02": 10.50
}

# Ask the user for a product ID
product_id = input("Enter product ID: ")

# Check if the product exists
if product_id in products:
    print(product_id, products[product_id])
else:
    print(product_id, "not found")
```

```python
name_counts = {}

while True:
    name = input("Enter a name (type 'done' to stop): ")

    if name.lower() == "done":
        break

    if name not in name_counts:
        name_counts[name] = 1
    else:
        name_counts[name] += 1

print("\nFinal Dictionary:")
print(name_counts)
```
