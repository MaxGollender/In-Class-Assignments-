# In-Class Assignment 1


```python
# Income Statement Program

# Get sales input from user
sales = float(input("Enter sales ($): "))

# Calculate Cost of Goods Sold (40% of Sales)
cost_of_goods_sold = sales * 0.40

# Calculate Gross Profit
gross_profit = sales - cost_of_goods_sold

# Get tax rate from user
tax_rate = float(input("Enter tax rate (%): "))

# Calculate Taxes and Net Income
taxes = gross_profit * (tax_rate / 100)
net_income = gross_profit - taxes

# Display the Income Statement
print("\n Income Statement ")
print(f"Sales:              ${sales:,.2f}")
print(f"Cost of Goods Sold: ${cost_of_goods_sold:,.2f}")
print(f"Gross Profit:       ${gross_profit:,.2f}")
print(f"Taxes:              ${taxes:,.2f}")
print(f"Net Income:         ${net_income:,.2f}")
```

# In-Class Assignment 2

```python
# Simple Numerical Calculator

# Get user inputs
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
operator = input("Enter an operator (+, -, *, /): ")

# Perform calculation
if operator == "+":
    result = num1 + num2
elif operator == "-":
    result = num1 - num2
elif operator == "*":
    result = num1 * num2
elif operator == "/":
    if num2 != 0:
        result = num1 / num2
    else:
        print("Error: Division by zero is not allowed.")
        exit()
else:
    print("Invalid operator.")
    exit()

# Print result
print("Result:", result)
```

# In-Class Assignment 3

```python
base_fee = 5
hourly_rate = 2.50
min_fee = 10
max_fee = 20

for hours in range(1, 9):
    fee = base_fee + hourly_rate * hours

    # Apply minimum and maximum limits
    if fee < min_fee:
        fee = min_fee
    elif fee > max_fee:
        fee = max_fee

    print(hours, fee)
```

```python
import math

ticket_price = 10
base_attendees = 20
fixed_cost = 200

print("Ad$   Profit")

for ad in range(0, 201, 25):
    # Calculate additional attendees
    additional = 2 * math.sqrt(ad)

    # Total attendees
    total_attendees = base_attendees + additional

    # Revenue
    revenue = ticket_price * total_attendees

    # Profit
    profit = revenue - ad - fixed_cost

    # Print results
    print(f"{ad:.1f}  {profit:.1f}")
```
